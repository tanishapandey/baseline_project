```python
learning_rate = 0.000001
num_epochs = 700
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model = SkipConnectNet().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr = learning_rate, momentum = 0.9, weight_decay = 0.0001)
```

```python
train_input = torch.tensor(input_patch)
train_label = torch.tensor(output_patch)
```

```python
from torch.utils.data import TensorDataset, DataLoader, RandomSampler, SequentialSampler
batch_size = 120

train_data = TensorDataset(train_input, train_label)
train_sampler = RandomSampler(train_data)
train_dataloader = DataLoader(train_data, sampler=train_sampler, batch_size=batch_size)
```

```python
for epoch in range(num_epochs):
    for batch_id, (train_input, train_label) in enumerate(train_dataloader):
        optimizer.zero_grad() # eliminar gradientes acumulados
        train_input = torch.unsqueeze(train_input, 1)
        train_input = train_input.type(torch.FloatTensor)
        train_input = train_input.cuda().to(device)
        output, output_softmax = model(train_input)
        train_label = train_label.type(torch.LongTensor)
        train_label = train_label.cuda().to(device)
        loss = criterion(output_softmax, train_label)
        
        # Backward
        loss.backward()
        optimizer.step()
        
        if batch_id % 100 == 0:
            print('Epoch [{}/{}], Loss:{:.4f}'
                .format(epoch+1, num_epochs, loss.data))
```
